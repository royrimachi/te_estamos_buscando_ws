# Dependencies

```python
import pandas as pd
import re
```

# Read datasets

```python
TEB_encontrados = pd.read_csv('teestamosbuscando_2019_11_25_found_people_profiles.csv', encoding='latin-1')
TEB_encontrados
```

```python
auxDf = TEB_encontrados
```

```python
auxDf['eyes_color'] = (auxDf
                       .apply(lambda row: row['eyes_color'], axis=1)
                       .apply(lambda eye_color: str(eye_color))
                       .apply(lambda eye_color: eye_color.upper())
                       .apply(lambda eye_color: '0' if re.search('NEGR|OSCURO', eye_color) else eye_color)
                       .apply(lambda eye_color: '1' if re.search('MARR|CASTAÑO|PARD|CAF|CLAR', eye_color) else eye_color)
                       .apply(lambda eye_color: '2' if re.search('AMB|MIEL', eye_color) else eye_color)
                       .apply(lambda eye_color: '3' if re.search('AVELL', eye_color) else eye_color)
                       .apply(lambda eye_color: '4' if re.search('VERD', eye_color) else eye_color)
                       .apply(lambda eye_color: '5' if re.search('AZU', eye_color) else eye_color)
                       .apply(lambda eye_color: '6' if re.search('GRI', eye_color) else eye_color)
                       .apply(lambda eye_color: '-1' if re.search('NAN|[A-Z]', eye_color) else eye_color)
)
auxDf['eyes_color'].value_counts()
```

```python
auxDf['hair_color'] = (auxDf
                       .apply(lambda row: row['hair_color'], axis=1)
                       .apply(lambda hair_color: str(hair_color))
                       .apply(lambda hair_color: hair_color.upper())
                       .apply(lambda hair_color: '0' if re.search('NEGR|NGR', hair_color) else hair_color)
                       .apply(lambda hair_color: '1' if re.search('MARR|CASTAÑO|PARD|CARTAÑO|CAST', hair_color) else hair_color)
                       .apply(lambda hair_color: '2' if re.search('RUBI', hair_color) else hair_color)
                       .apply(lambda hair_color: '3' if re.search('ROJ', hair_color) else hair_color)
                       .apply(lambda hair_color: '4' if re.search('AZU', hair_color) else hair_color)
                       .apply(lambda hair_color: '5' if re.search('GRI|CANO|CAN', hair_color) else hair_color)
                       .apply(lambda hair_color: '0' if re.search('OSC', hair_color) else hair_color)
                       .apply(lambda hair_color: '-1' if re.search('NAN|[A-Z]', hair_color) else hair_color)
)
auxDf['hair_color'].value_counts()
```

```python
auxDf['gender'] = (auxDf
                       .apply(lambda row: row['gender'], axis=1)
                       .apply(lambda gender: str(gender))
                       .apply(lambda gender: gender.upper())
                       .apply(lambda gender: '0' if re.search('FEMENINO', gender) else gender)
                       .apply(lambda gender: '1' if re.search('MASCULINO', gender) else gender)
                       .apply(lambda gender: '-1' if re.search('NAN|[A-Z]', gender) else gender)
)
auxDf['gender'].value_counts()
```

```python
auxDf['mouth_size'] = (auxDf
                           .apply(lambda row: row['mouth_size'], axis=1)
                           .apply(lambda ms: str(ms))
                           .apply(lambda ms: ms.upper())
                           .apply(lambda ms: '0' if re.search('GRAND|GRUES|AMPL|ALARG|ANCH|CARN', ms) else ms)
                           .apply(lambda ms: '1' if re.search('MED|NORMAL|REGU', ms) else ms)
                           .apply(lambda ms: '2' if re.search('PEQUE|CHIC|DELG|FIN', ms) else ms)
                           .apply(lambda ms: '-1' if re.search('NAN|[A-Z]', ms) else ms)
                      )
auxDf['mouth_size'].value_counts()
```

```python
auxDf['nose'] = (auxDf
                     .apply(lambda row: row['nose'], axis=1)
                     .apply(lambda nose: str(nose))
                     .apply(lambda nose: nose.upper())
                     .apply(lambda nose: '0' if re.search('AGUIL|AGUL|ALARG|GRAND|GRUES|LARG|MOTO|ROMAN', nose) else nose)
                     .apply(lambda nose: '1' if re.search('ALT|ANGOST|ANGUL|BULBOS|NCAV|CONVEX|DERE|DESV|MED|MD|NORMAL|RECT|REDOND|REG|RESPINGA|SINUOS|PUNTI', nose) else nose)
                     .apply(lambda nose: '2' if re.search('AMPL|ANCH|CHAT|CHIC|CORT|FIN|ÑAT|OVAL|PEQUE|PERFIL|DELGA|ANGOST', nose) else nose)
                     .apply(lambda nose: '-1' if re.search('NAN|[A-Z]', nose) else nose)
                )
auxDf['nose'].value_counts()
```

```python
auxDf['race'] = (auxDf
                     .apply(lambda row: row['race'], axis=1)
                     .apply(lambda race: str(race))
                     .apply(lambda race: race.upper())
                     .apply(lambda race: '0' if re.search('MES|MEZ|ORIGIN', race) else race)
                     .apply(lambda race: '1' if re.search('BLANC|CAUCA|CLAR', race) else race)
                     .apply(lambda race: '2' if re.search('AMARILL|CANEL|TRIG|MORE|COBR', race) else race)
                     .apply(lambda race: '3' if re.search('NEGR', race) else race)
                     .apply(lambda race: '-1' if re.search('NAN|[A-Z]', race) else race)
                )
auxDf['race'].value_counts()
```
